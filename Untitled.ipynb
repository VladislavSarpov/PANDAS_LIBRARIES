Задание 1 Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies_df = pd.read_csv('movies.csv')

ratings_df = pd.read_csv('ratings.csv')

ratings_5_df = ratings_df[ratings_df['rating'] == 5.0]
#ratings_5_df

ratings_5_group_df = ratings_5_df[['movieId', 'rating']].groupby('movieId').count().reset_index()
ratings_5_group_df = ratings_5_group_df.sort_values(by='rating', ascending=False).reset_index(drop=True)
#ratings_5_group_df

mostRatedMovieId = ratings_5_group_df.loc[0, 'movieId']
#mostRatedMovieId

mostRatedMovieTitle = movies_df[movies_df['movieId'] == mostRatedMovieId].iloc[0].title
mostRatedMovieTitle
```

Задание 2 По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_df = pd.read_csv('power.csv')

countries = ['Lithuania', 'Latvia', 'Estonia']

filtered_power_df = power_df[power_df['country'].isin(countries)]
#filtered_power_df

categories = [4, 12, 21]
filtered_power_df = filtered_power_df[filtered_power_df['category'].isin(categories)]
#filtered_power_df

filtered_power_df = filtered_power_df[filtered_power_df['year'].between(2005, 2010)]
#filtered_power_df

filtered_power_df = filtered_power_df[filtered_power_df['quantity'] >= 0]
#filtered_power_df

filtered_power_df[['country', 'quantity']].groupby('country').sum().reset_index() # не обязательно. выводит потребление по указанным странам

filtered_power_df['quantity'].sum()
```
